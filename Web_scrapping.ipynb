```python
from bs4 import BeautifulSoup as bs
import pandas as pd
pd.set_option('display.max_colwidth',500)
import time
import requests
import random
```

```python
page = requests.get("http://quotes.toscrape.com/")
page
```

```python
soup = bs(page.content)
soup
```

```python
soup.find_all(class_='text')
```

```python
quotes = [i.text for i in soup.find_all(class_='text')]
quotes
```

```python
authors = [i.text for i in soup.find_all(class_='author')]
authors
```

```python
url = [f"http://quotes.toscrape.com/page/{i}/" for i in range(1,50)]
url
```

```python
rate = [i/10 for i in range(50)]
time.sleep(random.choice(rate))
```

```python
for i in url:
    p = requests.get(i)
    soup = bs(p.content)
    authors.extend([i.text for i in soup.find_all(class_='author')])
    quotes.extend([i.text for i in soup.find_all(class_='text')])
    if len(quotes)>=500:
        break
    time.sleep(random.choice(rate))
```

```python
df = pd.DataFrame()
df['Authors'] = authors
df['Quotes'] = quotes
```
